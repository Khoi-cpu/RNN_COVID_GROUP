```python
import pandas as pd     
```

```python
# Load dataset
csv_file_path = r'Datasets\Weekly_Influenza_Vaccination_Coverage_and_Intent_for_Vaccination__Overall__by_Selected_Demographics_and_Jurisdiction__Among_Adults_18_Years_and_Older_20241101.csv'
df = pd.read_csv(csv_file_path)
```

```python
# Create new dataset filtering for Race/Ethnicity, and "Received a vaccination" indicator label
filtered_df = df[
    (df['Demographic Level'] == 'Race/Ethnicity') & 
    (df['indicator_category_label'] == 'Received a vaccination')
    ].copy()  # Add .copy() here to ensure a new DataFrame


# Calculate the number of people who received a vaccination 
filtered_df['Yes Vaccination'] = (filtered_df['Estimates'] / 100) * filtered_df['Unweighted Sample Size']

# Calculate the number of people who did NOT receive a vaccination
filtered_df['No Vaccination'] = filtered_df['Unweighted Sample Size'] - filtered_df['Yes Vaccination']

# Remove rows where either Yes Vaccination or No Vaccination is 0
filtered_df = filtered_df[(filtered_df['Yes Vaccination'] != 0) | (filtered_df['No Vaccination'] != 0)]

# Create a pivot table to aggregate Yes and No vaccinations
pivot_df = filtered_df.pivot_table(
    index=['Geographic Level', 'Week_ending', 'Demographic Name'],
    values=['Yes Vaccination', 'No Vaccination'],
    aggfunc='sum'
).reset_index()

# Rename columns to match the requested format
pivot_df.rename(columns={
    'Demographic Name': 'Race/Ethnicity'

}, inplace=True)



# Display the resulting DataFrame
pivot_df

# Save the filtered DataFrame to a CSV file
pivot_df.to_csv("race_ethnicity.csv", index=False)
```
